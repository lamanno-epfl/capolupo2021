```python
import scanpy as sc
import sys
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# sc.logging.print_versions()
```

```python
# Read the data files
adata_integrated = sc.read('samples_fibroblasts_integrated.csv', cache=True)
adata_integrated = adata_integrated.transpose()
```

```python
# Basic scanpy analysis
sc.tl.pca(adata_integrated, svd_solver='arpack')
sc.pp.neighbors(adata_integrated, n_neighbors=10, n_pcs=40)
sc.tl.umap(adata_integrated)
sc.tl.louvain(adata_integrated)
```
